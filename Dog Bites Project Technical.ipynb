```python
from scipy import stats
from scipy.stats import normaltest
print(normaltest(nyc_bite_final))
print(normaltest(spayed_females))
print(normaltest(not_spayed_females))
print(normaltest(neutered_males))
print(normaltest(unneutered_males))
```

```python
from scipy.stats import kruskal
print(kruskal(spayed_females, neutered_males))
print(kruskal(not_spayed_females, unneutered_males))
print(kruskal(neutered_males, unneutered_males))
print(kruskal(spayed_females, not_spayed_females))
print(kruskal(total_males, total_females))
```

```python
def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    mean = np.mean(data)
    n = len(data)
    stderr = stats.sem(data)
    interval = stderr * stats.t.ppf((1 + confidence) / 2., n - 1)
    return (mean, mean - interval, mean + interval)

def report_confidence_interval(confidence_interval):
    location = "The mean lies within the interval ]{:.2}, {:.2}[".format(
    confidence_interval[1], confidence_interval[2])
    return location

neutered_m = confidence_interval(neutered_males)
print(neutered_m)
nmale_dogs = report_confidence_interval(neutered_m)
print(nmale_dogs)
unneutered_m = confidence_interval(unneutered_males)
print(unneutered_m)
unmale_dogs = report_confidence_interval(unneutered_m)
print(unmale_dogs)
spayed_f = confidence_interval(spayed_females)
print(spayed_f)
sfemale_dogs = report_confidence_interval(spayed_f)
print(sfemale_dogs)
notspayed_f = confidence_interval(not_spayed_females)
print(notspayed_f)
notsfemale_dogs = report_confidence_interval(notspayed_f)
print(notsfemale_dogs)
```
